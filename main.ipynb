```python
C= df1['vote_average'].mean()
print(C)
```

```python
m= df1['vote_count'].quantile(0.9)
m
```

```python
q_movies = df1.copy().loc[df1['vote_count'] >= m]
q_movies.shape
```

```python
def weightedrating(x,m=m,C=C):
    
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_movies['weight_score'] = q_movies.apply(weightedrating, axis=1)
```

```python
q_movies = q_movies.sort_values('weight_score', ascending=False)

#Print the top 20 movies
q_movies[['title', 'vote_count', 'vote_average', 'weight_score']].head(20)
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
# TfidfVectorizer - This is used to convert text data into numerical values
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
selected_features = ['genres','keywords','tagline','title']
print(selected_features)
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
def combine_features(row):
    try:
        return row['genres'] +" "+row['keywords']+" "+row["tagline"]+" "+row["title"]
    except:
        print("Error:", row)

movies_data["combined_features"] = movies_data.apply(combine_features,axis=1)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(movies_data["combined_features"])
```

```python
similarity = cosine_similarity(feature_vectors)
results_title = pd.DataFrame(similarity,index=movies_data.title,columns=movies_data.title)
results_pop = pd.DataFrame(similarity,index=movies_data.popularity,columns=movies_data.popularity)
```
